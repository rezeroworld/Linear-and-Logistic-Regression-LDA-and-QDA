**Question A:** Derive the form of the maximum likelihood estimator for the Fisher LDA model described above. Hint: you can re-use some of the tricks presented in class for the MLE of a multivariate Gaussian, but adapted to this setting. You can get inspiration from Section 7.2 in Mike’s book (which covers the case where $\Sigma$ is diagonal).

**Answer:** We have:

$l(\theta|D) = log(\prod_{n=1}^{N}p(y_n|\pi)p(x_n|y_n,\theta)$

$= \sum_{n=1}^{N} log(p(y_n|\pi)) + log(p(x_n|y_n,\theta))$

To compute $\pi_{MLE}$ we only need the first part of the equation since the second one does not get involved

$\pi_{MLE} = argmax_{\pi} l(\theta|D) = argmax_{\pi} \sum_{n=1}^{N} log(p(y_n|\pi)) = argmax_{\pi} \sum_{n=1}^{N} y_nlog(\pi) + (1-y_n)log(1-\pi)$

By deriving with respect to $\pi$ we get:

$\frac{dl}{d\pi} = \sum_{n=1}^{N} \frac{y_n -\pi}{\pi(1-\pi)}$ and by setting it equal to $0$ we have:

$\frac{dl}{d\pi} = 0-->\sum_{n=1}^{N} \frac{y_n -\pi}{\pi(1-\pi)} = 0 -->\pi_{MLE} = \frac{\sum_{n=1}^{N} y_n}{N}$

By analyzing the derivative we see that $if \pi \leq y_n$then the derivative is positive and if $\pi \geq y_n$ it is negative so it is clear that $\pi_{MLE}$ is a maximum of $l(\theta|D)$

To derive $\mu_0$ and $\mu_1$ we only use the second term for clear reasons so we have

$\mu_{1MLE} = argmax_{\mu_1} \sum_{n=1}^{N} log(p(x_n|y_n,\theta)) = argmax_{\mu_1} \sum_{n=1}^{N} log(p(x_n|y_n=1)^{y_n}) + log(p(x_n|y_n=0)^{(1-y_n)}) = argmax_{\mu_1} \sum_{n=1}^{N} y_nlog(p(x_n|y_n=1)) + (1-y_n)log(p(x_n|y_n=0))$

The second term will not appear in the derivative with resepct to $\mu_{1MLE}$ so we will have:

$\frac{dl}{d\mu_1} = \frac{d}{d\mu_1} \sum_{n=1}^{N} y_nlog(p(x_n|y_n=1)) = \sum_{n=1}^{N} y_nlog(\frac{1}{(2\pi)^2|\Sigma|^{\frac{1}{2}}}e^{-\frac{1}{2}(x_n -\mu_1)\top\Sigma^{-1}(x_n-\mu_1)}) = -\sum_{n=1}^{N}y_n\Sigma^{-1}(x_n-\mu_1)$

By setting it equal to $0$ we have 

$\frac{dl}{d\mu_1} = 0 --> -\sum_{n=1}^{N}y_n\Sigma^{-1}(x_n-\mu_1) = 0 -->  -\Sigma^{-1}\sum_{n=1}^{N}y_n(x_n-\mu_1) = 0$

$\mu_{1MLE} = \frac{\sum_{n=1}^{N}y_nx_n}{\sum_{n=1}^{N}y_n}$ 

We just multiply the equation by $\Sigma$ to get rid of the inverse. So by looking back at the the derivative we see that the function is concave then it is a maximum. (just a reminder $x_n$ is a 2D-vector) 

By doing the same thing for $\mu_0$ we have 

$\frac{dl}{d\mu_0} = -\sum_{n=1}^{N} (1-y_n)\Sigma^{-1}(x_n-\mu_0)$

By setting it to $0$ we have

$\frac{dl}{d\mu_0} = 0 --> -\sum_{n=1}^{N}(1-y_n)\Sigma^{-1}(x_n-\mu_0) = 0 -->  -\Sigma^{-1}\sum_{n=1}^{N}(1-y_n)(x_n-\mu_0) = 0$

$\mu_{0MLE} = \frac{\sum_{n=1}^{N}(1-y_n)x_n}{\sum_{n=1}^{N}(1-y_n)}$

So now we need to find $\Sigma_{MLE}$, let's do it

$l(\theta|D) = \sum_{n=1}^{N} y_nlog(\frac{1}{(2\pi)^2|\Sigma|^{\frac{1}{2}}}e^{-\frac{1}{2}(x_n-\mu_1)\top\Sigma^{-1}(x_n-\mu_1)}) + (1-y_n)log(\frac{1}{(2\pi)^2|\Sigma|^{\frac{1}{2}}}e^{-\frac{1}{2}(x_n-\mu_0)\top\Sigma^{-1}(x_n-\mu_0)})$

$\frac{dl}{d\Sigma} = -\frac{1}{2}\sum_{n=1}^{N} \Sigma^{-1} - \Sigma^{-1}A\Sigma^{-1}$

with $A = y_n(x_n-\mu_1)(x_n-\mu_1)\top + (1-y_n)(x_n-\mu_0)(x_n-\mu_0)\top$

By setting it equal to $0$ we get:

$\Sigma_{MLE} = \frac{\sum_{n=1}^{N}A}{N}$ and because the function is concave we have that it is a maximum.

**Question B:** What is the form of the conditional distribution $p(y = 1|x)$? Compare with the form of logistic regression

**Answer:** 

$P(y=1|x) = \frac{p(x|y=1)p(y=1)}{p(x)} = \frac{p(x|y=1)p(y=1)}{p(x|y=1)p(y=1) + p(x|y=0)p(y=0)} = \frac{1}{1+\frac{p(y=0)p(x|y=0)}{p(y=1)p(x|y=1)}} = \frac{1}{1 + e^{-(\frac{1}{2}(x-\mu_0)\top\Sigma^{-1}(x-\mu_0) - \frac{1}{2}(x-\mu_1)\top\Sigma^{-1}(x-\mu_1) + log\frac{\pi}{1-\pi})}} = \frac{1}{1+e^{-(\beta\top x + \gamma)}} = \sigma(\beta\top x+ \gamma)$

So we can say that the conditional distribution has the form of the logistic regression with parameters $\beta$ and $\gamma$

```python
def dummy_model(X, y):
    
    """
    This is a dummy model to show you how your solution is expected to look like
        Inputs:
            X: [nx2] matrix of inputs
            Y: [n] vector of labels
        
        Returns:
            dummy_classifier: dummy classification function
    """
    
    # Just computing some means and standard deviations, nothing fancy
    mu0, mu1 = X[:, 0].mean(), X[:, 1].mean()
    s0, s1 = X[:, 0].std(), X[:, 1].std()
    
    
    def dummy_classifier(x0, x1):
        # Note how we use the variables mu0, mu1, s0 and s1 computed from the data
        # to define this function. Their values get stored in the definition
        # of the function itself, meaning that we don't have to store them explicitly.
        return 0.5 * (1 + np.tanh((x0 - mu0)**2/s0 - (x1 - mu1)**2/s1))
    
    return dummy_classifier
```

```python
def sigmoid(x):
    #return 1 / (1 + np.exp(-x))
    return special.expit(x)

def pdf_multivariate_gauss(x, mu0, mu1, sigma0,sigma1, pi0, pi1):
    res = (1./2) * ((x-mu0).dot(np.linalg.inv(sigma0))).dot((x-mu0)) - (1./2) * ((x-mu1).dot(np.linalg.inv(sigma1))).dot((x-mu1)) - np.log(pi0/pi1)
    if res>0:
        return 1 / (1 + np.exp(-res))
    else:
        return np.exp(res) / (1 + np.exp(res))

def gaussian(x,mu,variance):
    return (1/np.sqrt(2*np.pi*variance))*np.exp(-((x-mu)**2)/(2*variance))
```

```python
def LDA(X, y):
    """
    Estimates the parameters of an LDA model
        
        Inputs:
            X: [nx2] matrix of inputs
            Y: [n] vector of labels
        
        Returns:
            LDA_classifier: function taking two scalar inputs implementing the
                            estimated p(y=1|x)
    """
    # TODO: parameter estimation goes here
    train_inputs = X
    train_labels = y
    nclasses = len(np.unique(train_labels))
    nfeatures = train_inputs.shape[1]
    class_indexes = []
    for i in range(nclasses):
        class_indexes.append(np.argwhere(train_labels == i))

    class_0_mean, class_1_mean = np.empty((nclasses, nfeatures)), np.empty((nclasses, nfeatures))
    SW_0, SW_1, SW = np.zeros((nclasses, nfeatures, nfeatures)), np.zeros((nclasses, nfeatures, nfeatures)), np.zeros((nclasses, nfeatures, nfeatures))
    W = np.zeros((nclasses, nfeatures, 1))
    W0 = np.zeros((nclasses))

    class_0_mean = np.mean(train_inputs[class_indexes[0]], axis=0)
    class_1_mean = np.mean(train_inputs[class_indexes[1]], axis=0)
    between_class_0 = np.subtract(train_inputs[class_indexes[0]].reshape(-1, nfeatures), 
                                 class_0_mean)
    SW_0 = between_class_0.T.dot(between_class_0)
    between_class_1 = np.subtract(train_inputs[class_indexes[1]].reshape(-1, nfeatures), class_1_mean)
    SW_1 = between_class_1.T.dot(between_class_1)
    SW = SW_0 + SW_1
    W = np.dot(np.linalg.pinv(SW), (class_1_mean - class_0_mean).squeeze().T)
    W0 = -0.5 * np.dot(W.T, (class_0_mean + class_1_mean).squeeze().T)

    def LDA_classifier(x0, x1):
         # TODO: use the variables estimated above to define p(y=1|(x0, x1))
        x = np.array([x0,x1])
        LDA_projections = np.dot(train_inputs, W) + W0
        mu0 = np.mean(LDA_projections[train_labels==0])
        mu1 = np.mean(LDA_projections[train_labels==1])
        variance = np.sum(train_labels*(LDA_projections - mu1)**2 + (1-train_labels)*(LDA_projections - mu0)**2) / train_labels.shape
        prior = np.mean(train_labels==1)
        LDA_projections_test = np.dot(x, W) + W0
        raw_prediction = (gaussian(LDA_projections_test, mu1, variance)*prior) / (gaussian(LDA_projections_test, mu1, variance)*prior + gaussian(LDA_projections_test, mu0, variance)*(1-prior))
        return raw_prediction
            
    return LDA_classifier


def LogReg(X, y):
    """
    Estimates the parameters of a Logistic Regression model
        
        Inputs:
            X: [nx2] matrix of inputs
            Y: [n] vector of labels
        
        Returns:
            LogReg_classifier: function taking two scalar inputs implementing the
                               estimated p(y=1|x)
    """
    # TODO: parameter estimation goes here
    W = np.random.normal(size=(X.shape[1]))

    train_inputs = np.append(X, np.ones_like(y).reshape(-1,1), axis=1)
    W = np.zeros(train_inputs.shape[1])

    W_1 = W - 10

    while np.linalg.norm(W_1 - W) >= 1:
        alpha = sigmoid(np.dot(W,train_inputs.T))
        B_ = np.dot(alpha.reshape(-1,1), (1 - alpha).reshape(1,-1))
        B = np.diag(np.diag(B_))
        g = np.dot(train_inputs.T, alpha - y)
        H = np.dot(train_inputs.T, np.dot(B,train_inputs))
        W_1 = W
        W = W - np.linalg.lstsq(H, g, rcond=None)[0]
    
    def LogReg_classifier(x0, x1):
         # TODO: use the variables estimated above to define p(y=1|(x0, x1))
        x = x = np.array([x0,x1,1])
        raw_prediction = sigmoid(np.dot(W,x.T))
        return raw_prediction
    
    return LogReg_classifier


def LinReg(X, y):
    """
    Estimates the parameters of a Linear Regression model
        
        Inputs:
            X: [nx2] matrix of inputs
            Y: [n] vector of labels
        
        Returns:
            LinReg_classifier: function taking two scalar inputs implementing the
                               estimated f(x)
    """
    # TODO: parameter estimation goes here
    train_inputs = X
    train_labels = y
    W = np.linalg.lstsq(np.append(train_inputs, np.ones_like(y).reshape(-1,1), axis=1), train_labels, rcond=None)[0]

    def LinReg_classifier(x0, x1):
         # TODO: use the variables estimated above to implement f(x0, x1)
        x = np.array([x0,x1,1])
        raw_prediction = np.dot(x, W)
        prediction = np.where(raw_prediction < 0.5, 0, 1)        
        return prediction
    
    return LinReg_classifier


def QDA(X, y):
    """
    Estimates the parameters of a QDA model
        
        Inputs:
            X: [nx2] matrix of inputs
            Y: [n] vector of labels
        
        Returns:
            QDA_classifier: function taking two scalar inputs implementing the
                            estimated p(y=1|x)
    """
    # TODO: parameter estimation goes here
    train_inputs = X
    train_labels = y
    mu_0 = np.mean(train_inputs[train_labels==0], axis=0)
    mu_1 = np.mean(train_inputs[train_labels==1], axis=0)
            
    nfeatures = 2
    class_0_mean = mu_0
    class_1_mean = mu_1
    
    between_class_0 = np.subtract(train_inputs[train_labels==0].reshape(-1, nfeatures), class_0_mean)
    sigma_0 = between_class_0.T.dot(between_class_0)
    between_class_1 = np.subtract(train_inputs[train_labels==1].reshape(-1, nfeatures), class_1_mean)
    sigma_1 = between_class_1.T.dot(between_class_1)
    
    pi_0 = np.mean(train_labels==0)
    pi_1 = 1 - np.mean(train_labels==0)
    
    sigma_1 = sigma_1 / np.sum(train_labels)
    sigma_0 = sigma_0 / np.sum(1-train_labels)
    
    def QDA_classifier(x0, x1):
         # TODO: use the variables estimated above to implement p(y=1|(x0, x1))
        x = np.array([x0,x1])
        p = pdf_multivariate_gauss(x, mu_0, mu_1, sigma_0,sigma_1, pi_0, pi_1)
        return p
    
    return QDA_classifier
```

### Code for plotting results

```python
# ----------------------------------------------------------------------- #
#                      Do NOT modify this cell
# ----------------------------------------------------------------------- #

def pointwise_eval(f, umatrix, vmatrix):
    """
    Evaluates the function f over a grid given by the matrices u and V
        
        Inputs:
            f: [callable] function (of two scalar inputs) to be evaluated
            u: [n x m] matrix of values for the first input
            v: [n x m] matrix of values for the second input
                    
        Returns:
            z: [n x m] matrix of function values f(u_ij, v_ij) for i \in [1..n], j \in [1..m]
    """
    n, m = umatrix.shape
    z = np.zeros((n, m))
    for i in range(n):
        for j in range(m):
          z[i, j] = f(umatrix[i, j], vmatrix[i,j])
    return z
    
def plot_results(X, y, classifiers):
    """
    Displays the behavior of the several classifiers via a 2x2 subplot.
        
        Inputs:
            X: [nx2] matrix of inputs
            Y: [n] vector of labels
            classifiers: [4] list of functions in EXACTLT the following order:
                         ["LDA", "LogReg", "LinReg", "QDA"]
                         
        Output:
            2x2 subplot with showing a scatter plot of the data, a contour plot 
            for each classifier function and the level set f(x)=0.5 in black. 
        
    """
    
    assert len(classifiers) == 4
    
    labels = ["LDA", "LogReg", "LinReg", "QDA"]
    
    # Find range in each dimension and setup a meshgrid
    x0_list = np.linspace(X[:, 0].min(), X[:, 0].max(), 250)
    x1_list = np.linspace(X[:, 1].min(), X[:, 1].max(), 250)
    x0_v, x1_v = np.meshgrid(x0_list, x1_list, sparse=False, indexing='ij')
    
    # Create 2x2 subplot
    fig, axs = plt.subplots(2, 2, figsize=(9, 9))
    (ax1, ax2), (ax3, ax4) = axs
    
    for ix, ax in enumerate([ax1, ax2, ax3, ax4]):
        
        # Take an individual classifier
        f = classifiers[ix]
        
        if not f is None:
            
            # Evaluate over the meshgrid and plot contour lines
            # Z = f(x0_v, x1_v) # This is old code!!
            Z = pointwise_eval(f, x0_v, x1_v)
            contours = ax.contourf(x0_v, x1_v, Z, 5, cmap='RdBu', alpha=0.2);
            ax.clabel(contours, inline=True, fontsize=10)
            ax.contour(x0_v, x1_v, Z, [0.5], colors='black');

            # Plot the labelled data on top
            ax.scatter(X[y==0, 0], X[y==0, 1], marker="x", c='r');
            ax.scatter(X[y==1, 0], X[y==1, 1], marker="o", facecolors='none', edgecolors='b');
            ax.set_title(labels[ix])
```
